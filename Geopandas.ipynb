```python
#porfavor ese punto cree una serie de puntos para compararlos con la serie de mi json

universidad = gpd.GeoSeries(universidad, crs=4326) #geoseries = serie de valores de datos geograficos
universidad
```

##Agregación de Datos

```python
import pandas as pd
```

```python
poblacion = pd.read_csv("localidadpoblacion.csv", encoding='latin1', delimiter=';') #encoding es pa decirle que hay caracteres en español y delimiter ;
poblacion.head()
#para que cuadre este dataframe con el otro toca hacer limpieza de datos (coincider Usaquén con USAQUEN del otro dataframe)
```

```python
#como saco el registro de los datos del año 2024?
#Toca realizar un filtro en el dataframe

poblacion_2024 = poblacion[poblacion['Año'] == 2024]
poblacion_2024.head()
```

```python
#toca agrupar todos los valores de la misma localidad, se hace internamente para luego operar en ellos
grupos = poblacion_2024.groupby('Localidad')
print(grupos)
```

```python
#sumamos cada localidad
total_poblacion_localidad=grupos["Población"].sum()
total_poblacion_localidad
```

```python
#esta serie lo convertimos en un dataframe -> reset_index() para convertirlo en dataframe

df_poblacion_localidad = total_poblacion_localidad.reset_index()
df_poblacion_localidad
```

```python
#el problema es que esta en minisculas y tildes, toca decirle que la serie sea mayuscula str.upper

df_poblacion_localidad['Localidad']=df_poblacion_localidad['Localidad'].str.upper()
df_poblacion_localidad
```

```python
# TILDES

def replace_accents(input_str):
  replacements ={
      'Á':'A',
      'É':'E',
      'Í':'I',
      'Ó':'O',
      'Ú':'U',
  }
  for accented_char, unaccented_char in replacements.items():
    input_str = input_str.replace(accented_char, unaccented_char)
  return input_str


df_poblacion_localidad['Localidad'] = df_poblacion_localidad['Localidad'].apply(replace_accents)
print (df_poblacion_localidad)
```

```python
marged_df = gdf_localidades.merge(df_poblacion_localidad, left_on='LocNombre', right_on='Localidad')
marged_df.drop('Localidad', axis=1, inplace=True)
marged_df
```

```python
import matplotlib.pyplot as plt
```

```python
mapa_poblacion = marged_df.plot(
    column="Población",
    figsize=(15, 15),
    edgecolor='black',  # contorno
    cmap='Blues',  # color para la población
    legend=True,  # crear leyenda
)
mapa_poblacion.set_title("Mapa de Población")

mapa_poblacion
```
